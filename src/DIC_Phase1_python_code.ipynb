```python
import pandas as pd
import numpy as np

df = pd.read_csv("diabetes_prediction_dataset.csv")

# Display the first few rows of the DataFrame
print(df.head())
print(df.columns)
```

1.DROPPING DUPLICATES

```python
duplicates = df[df.duplicated()]

df.drop_duplicates(inplace = True)
df.head()
df.count()
```

2. HANDLING NULL VALUES

```python

missing_values = df.isnull().sum()
df = df.dropna()
# df.head()
df.count()
```

3. REMOVING UNWANTED WHITE SPACES

```python
df['gender'] = df['gender'].str.strip()
df['smoking_history'] = df['smoking_history'].str.strip()
df.head()
```

4. CONVERTING DATA TYPES

```python
df['gender'] = df['gender'].astype('category')
df['smoking_history'] = df['smoking_history'].astype('category')

print(df.dtypes)
```

5. CREATING NEW FEATURES

I. Seperating Age groups into Child, Young Adult, Adult, Middle-aged and Senior

```python

df['age_group'] = pd.cut(df['age'],
                         bins=[0, 18, 35, 50, 65, 100],
                         labels=['Child', 'Young Adult', 'Adult', 'Middle-aged', 'Senior'])

print(df[['age', 'age_group']].head())
```

II. Dividing BMI Into categories like Underweight, Normal weight, Overweight, Obese

```python

df['bmi_category'] = pd.cut(df['bmi'],
                            bins=[0, 18.5, 24.9, 29.9, 100],
                            labels=['Underweight', 'Normal weight', 'Overweight', 'Obese'])

print(df[['bmi', 'bmi_category']].head())
```

III. Creating Interaction features (helps to find the combined effect)

```python

df['hypertension_heart_disease'] = df['hypertension'] * df['heart_disease']

print(df[['hypertension', 'heart_disease', 'hypertension_heart_disease']].head())
```

IV. Calculating health risk score to know the patients overall health risk

```python

df['health_risk_score'] = df['hypertension'] + df['heart_disease'] + (df['age'] > 50).astype(int)

print(df[['age', 'hypertension', 'heart_disease', 'health_risk_score']].head())
```

V. Calculating HbA1c to glucose ratio to know relationship between long-term blood sugar control and the current glucose level

```python

df['HbA1c_to_glucose_ratio'] = df['HbA1c_level'] / df['blood_glucose_level']

print(df[['HbA1c_level', 'blood_glucose_level', 'HbA1c_to_glucose_ratio']].head())
```

VI. Flagging for high glucose levels to categorize patients into high and low blood glucose levels

```python

df['high_glucose'] = (df['blood_glucose_level'] > 150).astype(int)
print(df[['blood_glucose_level', 'high_glucose']].head())
```

8. OUTLIER DETECTION AND HANDLING.

```python
from scipy import stats
import numpy as np

# Select numerical columns for outlier detection
numerical_columns = ['age', 'bmi', 'HbA1c_level', 'blood_glucose_level',
                     'hypertension_heart_disease', 'health_risk_score',
                     'HbA1c_to_glucose_ratio']

# Calculate Z-scores for the selected columns
z_scores = np.abs(stats.zscore(df[numerical_columns]))

# Set a threshold for outliers (commonly 3)
threshold = 3

# Identify outliers (rows where Z-score exceeds threshold)
outliers = np.where(z_scores > threshold)

# Remove rows that contain outliers
df_cleaned_zscore = df[(z_scores < threshold).all(axis=1)]

# Check the shape of the DataFrame before and after
print(f"Original number of rows: {df.shape[0]}")
print(f"Number of rows after removing outliers using Z-Score: {df_cleaned_zscore.shape[0]}")
```

The boxplot at the top offers a quick visual summary, showing the distribution's spread and central values like the median, which divides the data into two halves. This gives a clear picture of where most of the data points lie and highlights those that are exceptions or potentially erroneous, standing outside the normal range.

Directly beneath the boxplot, the histogram offers a detailed breakdown of the BMI values, grouping them into bins. This allows us to observe the shape of the distribution—whether it is symmetric, skewed, or perhaps shows multiple peaks, each of which can tell a different story about the population's health.

Overlaid on the histogram is a normal distribution curve based on the mean and standard deviation calculated from the data. This curve is useful for comparing the actual distribution of BMI to a theoretical normal distribution, highlighting deviations such as skewness or unusual variability.

The display of key statistics like the mean, median, and standard deviation alongside the visualizations provides a snapshot of the data’s central tendency and variability. This not only helps in understanding what a typical BMI might be but also how varied the BMI values are across the population.

Additionally, the script marks the bounds at three standard deviations from the mean, which helps identify extreme values considered statistical outliers. These are the values that are significantly different from the norm and could indicate special cases or issues with data collection.

By comparing the mean and median, and observing how the data points are distributed around these values, we can assess the skewness of the BMI distribution. A significant difference between these two measures might suggest a skewed distribution, which can have implications for how we interpret the overall health of the population.



```python
from scipy.stats import norm
import warnings
warnings.filterwarnings('ignore')

fig, (box_ax, hist_ax) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.5, 1)})

mean_value = np.mean(df['bmi'])
std_dev = np.std(df['bmi'])
median_value = np.median(df['bmi'])
upper_bound = mean_value + (3 * std_dev)
lower_bound = mean_value - (3 * std_dev)
print(f"Upper Bound: {upper_bound}")
print(f"Lower Bound: {lower_bound}")

sns.boxplot(x=df["bmi"], ax=box_ax)
box_ax.axvline(mean_value, color='b', linestyle='--')


sns.distplot(df["bmi"], ax=hist_ax,kde=False,norm_hist=True)
hist_ax.axvline(lower_bound, color='r', linestyle='--')
hist_ax.axvline(mean_value, color='b', linestyle='-')
hist_ax.axvline(upper_limit, color='r', linestyle='--')



plt.legend({'Mean':mean_value})

box_ax.set(xlabel='')

x = np.linspace(mean_value - 3*std_dev, mean_value + 3*std_dev, 100)
plt.plot(x, norm.pdf(x, mean_value, std_dev))
plt.legend({"lower_limit":lower_bound,'Mean':mean_value,'upper_limit':upper_limit})
print("Mean Value : ",mean_value)
print("Median Value : ",median_value)
print("1st Standard Deviation : ",std_dev)

plt.show()
```


The summary statistics:

Mean: Approximated to 0, indicating a roughly centered BMI distribution.

Standard Deviation (std): 1, showing a standard normal distribution.

Upper & Lower Limits: ±3 from the mean, used to detect outliers.
